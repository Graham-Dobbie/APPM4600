```python
# import libraries
import numpy as np

# define routines
def bisection(f,a,b,tol):
    
#    Inputs:
#     f,a,b       - function and endpoints of initial interval
#      tol  - bisection stops when interval length < tol

#    Returns:
#      astar - approximation of root
#      ier   - error message
#            - ier = 1 => Failed
#            - ier = 0 == success

#     first verify there is a root we can find in the interval 

    fa = f(a)
    fb = f(b);

    err_arr = np.array([])
    
    if (fa*fb>0):
       ier = 1
       astar = a
       return [astar, ier, count, err_arr]

#   verify end points are not a root 
    if (fa == 0):
      astar = a
      ier =0
      return [astar, ier, count, err_arr]

    if (fb ==0):
      astar = b
      ier = 0
      return [astar, ier, count, err_arr]

    count = 0
    d = 0.5*(a+b)
    while (abs(d-a)> tol):
      fd = f(d)
      if (fd ==0):
        astar = d
        ier = 0
        return [astar, ier, count, err_arr]
      if (fa*fd<0):
         b = d
      else: 
        a = d
        fa = fd
      d = 0.5*(a+b)
      count = count +1
      err_arr = np.append(err_arr, d)
      
      # print('abs(d-a) = ', abs(d-a))
      
    astar = d
    ier = 0
    return [astar, ier, count, err_arr]
```

```python
def driver_bisection(f,a,b,tol,true_r = None):

    [astar,ier, count, err] = bisection(f,a,b,tol)
    if true_r is None:
        true_r = astar
        
    n = np.arange(err.size)
    plt.plot(n,np.log(abs(err-true_r)))
    plt.show()
    print('the approximate root is',astar)
    print('the error message reads:',ier)
    print('f(astar) =', f(astar))
    print('iters =', count)
    
    
```

```python
def fixed_point(f,x0,TOL,Nmax):
    x_arr = np.array([])
    x_n = np.inf
    x_np1 = x0
    iterations = 0

    while( iterations < Nmax):
        x_n = x_np1
        x_arr = np.append(x_arr, x_n)
        x_np1 = f(x_n)
        iterations += 1
        if (abs(x_n - x_np1) < TOL):
            return x_n, iterations, x_arr
    # print(err_arr)
    raise Exception("GAD DAMMIT you didn't find the root")
    return
```

```python
fix_point_func = lambda x : -np.sin(2*x) + 5*x/4 + -3/4
fixed_point(fix_point_func, 2, 1e-10, 100)
```

```python
x_converges = np.array([])
for i in np.arange(-2,6,0.1):
    try:
        x_final, _, _ = fixed_point(fix_point_func, i, 1e-11, 100)
        x_converges = np.append(x_converges,x_final)
    except:
        pass


        
    
```
