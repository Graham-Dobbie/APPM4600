```python
def wavy_circles(R,dr, f, p):
    theta = np.linspace(0, 2*np.pi,200)
    x = R*(1 + dr*np.sin(f*theta + p))*np.cos(theta)
    y = R*(1 + dr*np.sin(f*theta + p))*np.sin(theta)
    return x,y
```

```python
[x,y] = wavy_circles(1.2,0.1, 15, 0)
plt.plot(x,y)
plt.show()
```

```python
x = []
y = []
for i in range(10):
    x_new, y_new = wavy_circles(i, 0.05, 2+i, uniform(0,2))
    for j in range(len(x_new)):
        x.append(x_new[j])
        y.append(y_new[j])
plt.plot(x,y)
plt.show()
```
