```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
def eval_legendre(x,n):
    p = np.zeros(n+1)
    p[0] = 1
    p[1] = x

    for i in range(1, n):
        p[i+1] = 1/(i+1)*((2*i + 1)*x*p[i] - i*p[i-1])

    return p

def eval_legendre_nd(x,n):
    p_arr = np.zeros([len(x), n+1])
    for i in range(len(x)):
        p_arr[i] = eval_legendre(x[i], n)

    return p_arr
    

def L2_dot(f1, f2, a, b, w = None,):
    
    func = lambda x: f1(x)*f2(x)
    if (w != None):
        func = lambda x: f1(x)*f2(x)*w(x)

    (prod, err) =  quad(func, a, b)
    return prod
```
