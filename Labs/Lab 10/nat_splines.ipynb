```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5
smallest_cube = -10
largest_cube = 10
```

```python

def get_cube(x,N ,smallest_cube, largest_cube):
    rng_cubes = (largest_cube - smallest_cube)
    cube_term_list = rng_cubes*(np.random.random((N,)) - 0.5)

    e_vals = np.zeros(len(x))
    for j in range(len(x)):
        for i in range(N-1):
            if ((x[j] >= i) and(x[j] <= i+1)):
                slope = cube_term_list[i+1] - cube_term_list[i]
                e_vals[j] =  slope*(x[j]-i) + cube_term_list[i]
    return e_vals
                
```

```python
cube_func = lambda x : get_cube(x, N, smallest_cube, largest_cube)

x = np.linspace(0.5,3.5, 100)

plt.plot(x, cube_func(x))
plt.show()
```
