```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def comp_simpsonz(a, b, f, N):
    points = 2*int(N/2)+1
    intervals = np.linspace(a,b,points)
    f_evals = f(intervals)
    w = np.ones(points)
    for i in range(0,points):
        if (i%2):
            w[i] = 4
        else:
            w[i] = 2
    w[0] = 1
    w[-1] = 1

    
    # print(f_evals)
    int_f = np.dot(f_evals, w)
    return (b-a)/(3*(points-1)) * int_f

def comp_trapz(a, b, f, N):
    intervals = np.linspace(a,b,N+1)
    f_evals = f(intervals)
    w = np.ones(N+1)
    w[0] = 1/2
    w[-1] = 1/2

    int_f = np.dot(f_evals, w)
    return (b-a)/(N) * int_f
```

```python
N = 5
smallest_cube = -10
largest_cube = 10
```

```python

def get_cube(x,N ,smallest_cube, largest_cube):
    rng_cubes = (largest_cube - smallest_cube)
    cube_term_list = rng_cubes*(np.random.random((N,)) - 0.5)

    e_vals = np.zeros(len(x))
    for j in range(len(x)):
        for i in range(N-1):
            if ((x[j] >= i) and(x[j] <= i+1)):
                slope = cube_term_list[i+1] - cube_term_list[i]
                e_vals[j] =  slope*(x[j]-i) + cube_term_list[i]
    return e_vals
                
```

```python
cube_func = lambda x : get_cube(x, N, smallest_cube, largest_cube)

x = np.linspace(0.5,3.5, 100)

plt.plot(x, cube_func(x))
plt.show()
```

```python
def get_sqr(x,cube_func):
    min_x = min(x)
    max_x = max(x)
    N = 100
    n = np.linspace(min_x,max_x, N)

    int_cube = np.zeros(N)
    for i in range(1,N):
        # print( comp_simpsonz(n[i-1], n[i], cube_func, 10))
        int_cube[i] = comp_trapz(n[i-1], n[i], cube_func, 10)

    eval_int = np.interp(x, n, int_cube)
    return eval_int
```
